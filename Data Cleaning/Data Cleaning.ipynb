```python
# Making a DataFrame which is a dictionary of Lists

dict_1 = {'Name' : ['Manoj','Mukesh','Nidhi','Payal'],
          'Marks_100' : [80,40,60,90],
          'Age' : [22,24,27,24],
          'Team' : ['A','B','C','D'],
          'College' : ['DAV','MR','BSA','NCU']}

dataframe = pd.DataFrame(dict_1)
print(dataframe)
```

```python
# accessing rows using index

for index in dataframe.index:
    print(dataframe['Name'][index], dataframe['Age'][index])
```

```python
# accessing rows using loc

for index in dataframe.index:
    print(dataframe.loc[index, 'Name'], dataframe.loc[index, 'Age'])
```

```python
# accessing rows using loc

for index in range(len(dataframe)):
    print(dataframe.loc[index, 'Name'], dataframe.loc[index, 'Age'])
```

```python
# accessing rows using iloc

for index in range(len(dataframe)):
    print(dataframe.iloc[index, 0], dataframe.iloc[index, 2])
```

```python
# apply() method

dataframe.apply(lambda row: row['Name'] + " " + str(row['Age']), axis =1)
```

```python
# iterrows() method
# showing data as per the index(row)

for index, data in dataframe.iterrows():
    print("\n index:",index,"\n \n ",data)
```

```python
# iteritems() method

for column_name,column_data in dataframe.iteritems():
    print(column_name)
```

```python
# findall returns the list containing all the matches in a string

import re

string = '101:python 102:java 103:blockchain 104:crypto'

pattern1 = '\d'
pattern2 = '\d+'

result1 = re.findall(pattern1,string)
result2 = re.findall(pattern2,string)

print(result1)
print(result2)
```

```python
## search method returns the match object and the first occurence of the match

string = '101:python 102:java 103:blockchain 104:crypto'

pattern1 = '\d'
pattern2 = '\d+'

result1 = re.search(pattern1,string)
result2 = re.search(pattern2,string)

print(result1)
print(result2)
```

```python
# split method returns the list by splitting the string in accordance to input pattern

string = '101:python 102:java 103:blockchain 104:crypto'

pattern1 = '\d+'

result1 = re.split(pattern1,string)

print(result1)
```

```python
# we can also give order/length till which we want to split the string

n = 2 
string = '101:python 102:java 103:blockchain 104:crypto'

pattern1 = '\d+'

result1 = re.split(pattern1,string,n)

print(result1)
```
