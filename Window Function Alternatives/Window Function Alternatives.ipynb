```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv('employee.csv')
employee
```

```python
employee.info()
```

```python
employee.sort_values(by = ['dept_name', 'salary'])
```

```python
# select *, min(salary) over (partition by dept_name) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].transform(min)
```

```python
# select *, max(salary) over (partition by dept_name) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].transform(max)
```

```python
# select *, avg(salary) over (partition by dept_name) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].transform('mean')
```

```python
# select *, sum(salary) over (partition by dept_name) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].transform(sum)
```

```python
# select *, count(1) over (partition by dept_name) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].transform('count')
```

```python
# select *, min(salary) over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].cummin()
```

```python
# select *, max(salary) over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].cummax()
```

```python
# select *, sum(salary) over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].cumsum()
```

```python
# select *, row_number() over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name').cumcount()+1
```

```python
# select *, lag(salary) over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].shift(1)
```

```python
# select *, lead(salary) over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].shift(-1)
```

```python
# select *, row_number() over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].rank(method = 'first')
```

```python
# select *, rank() over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].rank(method = 'min')
```

```python
# select *, dense_rank() over (partition by dept_name order by salary) from employee

employee.sort_values(by = ['dept_name', 'salary']).groupby('dept_name')['salary'].rank(method = 'dense')
```
