```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
style.use('ggplot')
x = np.linspace(0,4*np.pi,1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(211)
plt.plot(x, y1, color = 'blue', label = 'Sine', linewidth = 2)
plt.legend()
plt.grid(True, color = 'black')
plt.title('Sine Vs. Cosine Plot', fontsize = 10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(212)
plt.plot(x, y2, color = 'orange', label = 'Cosine', linewidth = 2)
plt.legend()
plt.grid(True, color = 'black')
plt.title('Sine Vs. Cosine Plot', fontsize = 10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
